## Data Description

The dataset was obtained from a competition hosted on [Kaggle](https://www.kaggle.com/c/diabetic-retinopathy-detection). The images in the dataset is a large collection of high-resolution retina images taken under a wide variety of imaging conditions. Each subject provides two images, one for the left eye and one for the right. Images are named in the following convention: `<subjectID>_<left|right>.jpeg`(1_left.jpeg is the left eye of subject 1).

A clinician has rated the presence of diabetic retinopathy in these images on a scale of 0-4 according the following scale:

0 - No DR

1 - Mild

2 - Moderate

3 - Severe

4 - Proliferative DR

The images in the dataset come from a wide variety of models and type of cameras, which can affect the apperance of left vs. right. Some images are shown as one would see the retina anatomically(macula on the left, optic nerve on the right for the right eye). Others are shown as one would see them through a microscope condensing lens (i.e. inverted, as one sees in a typical eye exam). There are two ways to tell if an image in inverted:

* It is inverted if the macula(the small dark central area) is slightly higher than the midline through the optic nerve. If the macula is lower than the midline of the optic nerve, it's not inverted.

* If there is a notch on the side of the image then it's not inverted. If there is no notch then it is inverted.

### Preprocessing of dataset

As many different instruments or microscopes were used to obtain the retina images, the training images need to be preprocessed to create a consistent dataset. Images in the dataset and practical use come in a wide variety of dimensions, have noise and other artifacts, don't have good contrast and are not uniform. As neural networks and other learning algorithms expect images of similar nature, we need to create images of the same size.

#### Preprocessing Code

```shell
#!/bin/bash

# To run, this assumes that you are in the directory with the images
# unpacked into train/ and test/.  To run, it works best to use GNU
# parallel as
#
#   ls train/*.jpeg test/*.jpeg | parallel ./process.sh
#
# Otherwise, it also works to do a bash for loop, but this is slower.
#
#   for f in `ls train/*.jpeg test/*.jpeg`; do ./process.sh $f; done
#

# Size of the output image
size=512x512

in_file=$1
out_file=$1

com="convert $in_file \
-set colorspace HSL -channel lightness -equalize \
-set colorspace RGB -channel G -separate -depth 8 \
-convolve Gaussian \
-background black \
-fuzz 10% -trim +repage -resize $size \
-gravity center -background black -extent $size \
-resize $size \
processed/$out_file"

echo $com

$com
```

The techniques applied for image processing include:

1. Seperation of the green channel

2. Gaussian Filtering for noise removal

3. Trimming of the black space around the retina

4. Center the retina circle in the center of the image

5. Perfectly black background

6. Histogram equalization of the **light** channel in the HSL colorspace

Preprocessing of the images was done using ImageMagick which is a image processing tool for batch processing of images. It provides a command line interface to image processing which we used in the shell script above. Also we used GNU parallel to parallelize the processing of the images for faster processing.
